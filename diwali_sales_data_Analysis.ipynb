# Cleaning Datasets

```python
# remove null values
#df.dropna(inplace=True), it drop all complete row,so we don't procede with it

df['Amount']=df['Amount'].fillna(0)
```

```python
# chnage data type of "Amount" column from string to int type

df['Amount'] = df['Amount'].astype(int)
```

# Data Visualisation and Insights

```python
sns.countplot(x='Gender',data=df)
sns.set(rc={'figure.figsize':(6,5)})
plt.show()
```

```python
# to add data label and title in this plot

ax=sns.countplot(x='Gender',data=df,color="r")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()    
    
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

# From above plot and talbe we conclude that most of the buyers are female and even purchasing power of female are greater than male.

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')  # hue->show plot bseperately by gender,otherwise it shows total
sns.set(rc={'figure.figsize':(20,10)})               
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount vs age group

df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

# From above plot and table, we conclude that most of the buyers are of age group between 26-35 yrs male and famele.

```python
# top 10 states by orders

state_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
state_orders.head(10)
```

```python
state_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
state_orders.head(10)
sns.set(rc={'figure.figsize':(25,10)})               # to adjust plot size so that data visible
sns.barplot(data=state_orders,x='State',y='Orders')
```

```python
# total states by amount/sales
state_orders=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
state_orders.head(10)
sns.set(rc={'figure.figsize':(25,10)})               # to adjust plot size so that data visible
sns.barplot(data=state_orders,x='State',y='Amount',color="g")
```

# From above two plots,we conclude that most of the orders and sales or amount are from uttar pradesh(UP),Maharashtra and karnataka.

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(6,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
shadi=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
shadi   # 0 indicates married and 1 indicate unmarried
```

```python
shadi=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=shadi,x='Marital_Status',y='Amount',hue='Gender')
```

# From above plots we conclude that Most of the buyers and high purchasing power are married women.

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(30,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occ_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=occ_sales,x='Occupation',y='Amount',color="g")
```

# From above two plots, we conclude that Most of the buyers are working in IT,Healthcare and Aviation.

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(35,20)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
pro_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(data=pro_sales,x='Product_Category',y='Amount',color="magenta")
```

# From above plots, we conclude that most of the sold product are food,clothes and electronic.
